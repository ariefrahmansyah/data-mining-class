# k-means clustering

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data[:, [0,2]] # sepal length, petal length
target = iris.target
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

plt.scatter(x=data[:,0], y=data[:,1])

plt.xlabel('Sepal length')
plt.ylabel('Petal length')
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 2)

# training the k-means clustering in Python
km.fit(data)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=data[:,0], y=data[:,1], c=km.labels_)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
```

## Evaluation

```python
km.cluster_centers_
```

```python
km.inertia_
```

## Finding optimum clusters

```python
# Within Cluster Sum of Squares (WCSS) of clusters
wcss = [] 

# using for loop to iterates
for i in range(2, 15):
    #training k-means clustering in python on different clusters
    km = KMeans(n_clusters=i)
    km.fit(data)
    wcss.append(km.inertia_)
    
# ploting the elbow graph
plt.plot(range(2,15), wcss, 'og-')
plt.annotate('optimum clusters', xy=(5, 50000), xytext=(6, 100000), arrowprops=dict(facecolor='blue', shrink=0.05))

# labeling the data
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
```
